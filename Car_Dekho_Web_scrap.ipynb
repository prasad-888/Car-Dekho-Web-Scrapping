```python
def get_soup(url):
    """Takes URL and returns a soup object"""
    resp = requests.get(url,headers=headers)
    if resp.status_code == 200:
        return bs(resp.text)
    else: return None


def all_cars(url ='https://www.cardekho.com/maruti-suzuki-cars'):
  url1 = url
  price=[]
  car_name=[]
  type_car=[]
  b=[]
  soup1 = get_soup(url1)

  try:
    for i in soup1.find( 'section',class_="gsc_row gsc_container_hold heading BrandPagelist marginTop20").find_all('h3'):
      # print(i.text)
      car_name.append(i.text)


    for i in soup1.find( 'section',class_="gsc_row gsc_container_hold heading BrandPagelist marginTop20").find_all('div', class_='price'):
      # print(i.text)
      price.append(i.text)




    import numpy as np

    for i in  soup1.find('section',class_="gsc_row gsc_container_hold heading BrandPagelist marginTop20").find_all('div', class_='dotlist'):
      type_car.append([i.text for i in i.find_all('span')])

      b.append([i.text for i in i.find_all('span')])

    q=b[1::2]

    x = {'car_name':car_name, 'price':price, 'type':type_car[::2] , 'cc' : q }
    df = pd.DataFrame.from_dict(x, orient='index').T

    return df

  except:
    return None

# all_cars()
```

```python
all_df = pd.DataFrame()
for i in range(len(get_cars)):
  a12 = "https://www.cardekho.com" + get_cars[i]['href']
  temp_df = all_cars(a12)
  display(temp_df)
  all_df = pd.concat([all_df,temp_df])

all_df
```
